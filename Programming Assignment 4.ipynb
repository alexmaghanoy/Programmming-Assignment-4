## ECE BOARD EXAM PROBLEM

***PART 1: CREATE THE DATAFRAME***

```python
#import data set
df = pd.read_excel('board2.xlsx')
df
```

### PROBLEM 1.1

```python
# Filter the data by this conditions, where: the **Track** must be **Instrumentation**; the **Hometown** is **Luzon**
# The grade in **Electronics** must also be greater than 70
# Here, we will analyze the student's grade in GEAS based on these conditions.

Instru = df[(df['Track'] == 'Instrumentation') & 
            (df['Hometown'] == 'Luzon') & 
            (df['Electronics'] > 70)][['Name', 'GEAS', 'Electronics']]
Instru
```

### VISUALIZATION
Here is a sample bar graph for this data.

```python
plt.figure()
sns.barplot(x= "Name", y = "GEAS", data=Instru)
plt.title("Comparison of Grades in GEAS based on the Student's Track and Hometown")
plt.xlabel("Student")
```

### PROBLEM 1.2

```python
#this will compute the average grade of each student and create another column for the result
df['Average']=df[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)
```

```python
# Filter the data by this conditions, where: the **Gender** must be **Female**; the **Hometown** is **Mindanao**
# The **average grade** must also be greater than or equal to 55.
# Here, we will analyze the student's grade in **Electronics** based on these conditions.
Mindy = df[(df['Gender'] == 'Female') & 
            (df['Hometown'] == 'Mindanao') & 
            (df['Average'] >= 55)][['Name', 'Track', 'Electronics', 'Average']]
Mindy
```

### VISUALIZATION
Here is a sample box plot for this data.

```python
plt.figure()
sns.boxplot(x = "Track", y="Electronics", data = Mindy)
plt.title("Distribution of Grades (Electronics) across Different Tracks")
plt.xlabel("Tracks")
plt.ylabel("Electronics Grade")
```

```python
plt.figure()
sns.barplot(x="Name", y="Electronics", data=Mindy)
plt.title("Distribution of Grades (Electronics) of Each Students")
plt.xlabel("Student Number")
plt.ylabel("Electronics Grade")
```

***PART 2: CREATE THE VISUALIZATION OF DATA***

Create a visualization that shows how the different features contributes to average grade. 

Does chosen track in college, gender, or hometown contributes to a higher average score?

```python
#this will compute the average grade of each student and create another column for the result
df['Average']=df[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)
df
```

### USING BOXPLOT
- shows the distribution of grades for each group (track, gender, hometown) and highlight the median, quartiles, and potential outliers.

```python
#using boxplot, we will analyze the "Contribution of Chosen Track to the Average Grade"
plt.figure(figsize=(5,3))
sns.boxplot(x= "Track", y= "Average", data=df)
plt.title("Average Grade Distribution by Track")
plt.show()
```

```python
#using boxplot, we will analyze the "Contribution of Gender to the Average Grade"
plt.figure(figsize=(4,3))
sns.boxplot(x= "Gender", y= "Average", data=df)
plt.title("Average Grade Distribution by Gender")
plt.show()
```

```python
#using boxplot, we will analyze the "Contribution of Hometown to the Average Grade"
plt.figure(figsize=(5,3))
sns.boxplot(x= "Hometown", y= "Average", data=df)
plt.title("Average Grade Distribution by Hometown")
```

### USING HISTOGRAM
- best for visualizing the frequency distribution of a single numeric variable and see the overall shape of the grade distribution for each group. 

```python
plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
sns.histplot(df, x = "Average", hue = "Hometown", multiple = "stack")
plt.title("Average Grade Distribution by Hometown")

plt.subplot(2,2,2)
sns.histplot(df, x = "Average", hue = "Gender", multiple = "stack")
plt.title("Average Grade Distribution by Gender")

plt.subplot(2,2,3)
sns.histplot(df, x = "Average", hue = "Track", multiple = "stack")
plt.title("Average Grade Distribution by Track")

plt.tight_layout()
plt.show()
```

### USING VERTICAL BAR GRAPH
- provides a clear, straightforward way to visualize differences and make comparisons between different sets of data.

```python
#gender
plt.figure(figsize=(4,4))
plt.title("Average Grade Distribution by Gender")
sns.barplot(x="Gender", y="Average", data=df)
plt.show()
```

```python
#hometown
plt.figure(figsize=(5,3))
plt.title("Average Grade Distribution by Hometown")
sns.barplot(x="Hometown", y="Average", data=df)
plt.show()
```

```python
#track
plt.figure(figsize=(5,3))
plt.title("Average Grade Distribution by Track")
sns.barplot(x="Track", y="Average", data=df)
plt.show()
```
