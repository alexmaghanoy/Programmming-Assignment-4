```python
import seaborn as sns
```

### PROBLEM 1.1

```python
df = pd.read_excel('board2.xlsx')
df
```

```python
Instru = df[(df['Track'] == 'Instrumentation') & 
            (df['Hometown'] == 'Luzon') & 
            (df['Electronics'] > 70)][['Name', 'GEAS', 'Electronics']]
Instru
```

### VISUALIZATION

```python
plt.figure()
sns.barplot(x= "Name", y = "GEAS", data=Instru)
```

### PROBLEM 1.2

```python
df['Average']=df[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)
```

```python
Mindy = df[(df['Gender'] == 'Female') & 
            (df['Hometown'] == 'Mindanao') & 
            (df['Average'] >= 55)][['Name', 'Track', 'Electronics', 'Average']]
Mindy
```

### VISUALIZATION

```python
plt.figure()
sns.boxplot(x = "Track", y="Electronics", data = Mindy)
plt.title("Distribution of Grades (Electronics) across Different Tracks")
plt.xlabel("Tracks")
plt.ylabel("Electronics Grade")
```

```python
plt.figure()
sns.barplot(x="Name", y="Electronics", data=Mindy)
plt.title("Distribution of Grades (Electronics) of Individual Students")
plt.xlabel("Student Number")
plt.ylabel("Electronics Grade")
```
